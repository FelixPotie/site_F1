### Variables 

En programmation, vous utiliserez constemment des *variables*. Une variable est (comme en mathématiques ou en physique), une valeur auquel vous donnez un nom. Par exemple :

```python
a = 7 + 4

bonjour_je_suis_une_variable = "et je peux contenir du texte ! "
```

On peut ensuite afficher le contenu d'une variable avec la fonction `print`.

```python
print(a)
```

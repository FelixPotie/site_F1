!!! info Tri par comparaison
On a vu les algorithmes de tris suivants : tri par sélection, tri par insertion et tri bulle. Ces tris sont des tris appelés tris par comparaison. Ils utilisent un enchainement de comparaisons (`<` ou`>`), cependant il existe d'autre type de tris qui ne sont pas par comparaison. Nous allons en voir un aujourd'hui.
!!!

!!! note Version 1
Pour commencer, nous allons faire une version avec laquelle l'on veut trier des entiers naturels compris dans une borne `[0:m]`, où `m` est l'élément maximum de la liste. 

Le principe du tri par comptage est de :
1. Compter le nombre d'occurrences de chaque chiffre de la liste, on le stocke dans une liste `compteurs` de taille `m+1` où `compteurs[i]` contiendra le nombre d'occurrences de `i` dans la liste.
2. Initialiser une liste `liste_triee` vide 
2. Parcourir chaque élément `e` de `compteurs` et ajouter dans `liste_triee` le nombre `e` le nombre de fois qu'il faut (`compteurs[i]`).

Par exemple si on a la liste : \
`liste = [1,3,2,2,1]`\
alors va créer :\
`compteurs = [0,2,2,1]`\
et donc on aura :\
`liste_triee = [1,1,2,2,3]`.
!!!

!!! abstract `list.insert(indice, element)`
Le type list possède en fait déjà une fonction qui fait l'insertion d'un élément dans une liste à un certain indice.
```python
l = [5,1,6,9]
l.insert(1,2)

```
Sa complexité en pire cas est $O(n-i)$ avec $n =$ |liste|, car on doit d'écaler tous les éléments à partir de l'indice `i` vers la fin de la liste.
!!!

````python
l = [5,1,6,9]
l.insert(1,2)
print(l) #[5, 2, 1, 6, 9]
````

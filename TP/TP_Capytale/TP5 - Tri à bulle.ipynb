### Comparaison des 2 algorithmes 

!!! question Question 7
Comparer ces 2 fonctions avec n = 1 000 et 10 000.
!!!

```python
import time # Utilisation de la fonction perf_counter()
from random import randint # import du module random qui est renommé rd

n = 1000
liste_1 = [randint(0,100*n) for k in range(n)]
liste_2 = liste_1.copy()

# calcule du temps d'exécution de tri_bulle_1
temps_debut = time.perf_counter()
tri_bulle_1(liste_1)
temps_fin = time.perf_counter()
temps_execution_v1 = temps_fin - temps_debut

# calcule du temps d'exécution de tri_bulle_2
temps_debut = time.perf_counter()
tri_bulle_2(liste_2)
temps_fin = time.perf_counter()
temps_execution_v2 = temps_fin - temps_debut

print("Pour n =", n)
print("tri_bulle_1 à mis", temps_execution_v1, "s")
print("tri_bulle_2 à mis", temps_execution_v2, "s")
```

L'idée du tri cocktail consiste à alterner le sens du parcours.

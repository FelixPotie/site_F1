!!! info 
Le code suivant permet de tracer sur une figure un polygone dont on définit les `n` sommets en argument sous la forme d’une liste de couples de coordonnée `[[x_a, y_a], [x_b, y_b], ... ]`.

Par exemple avec les points `a,b,c` on dessine un triangle.
!!!

````python
from matplotlib import pyplot as plt


def polygone(liste_points) :
    x = []
    y = []
    for point in liste_points:
        x.append(point[0])
        y.append(point[1])
    
    plt.fill(x,y,'hotpink')
````

````python
a = [0,0]
b = [1,1]
c = [2,0]

plt.figure()
polygone([a,b,c])
plt.show()
````

## Exercice 3 - Cercles concentriques

Nous allons mettre en place un algorithme qui va réaliser une figure récursive où l’on affiche des
cercles concentriques de couleurs successivement noire puis blanche où le rayon diminue avec un pas
constant. Voici ci-dessous les images à obtenir pour différents cas :



Tous les cercles sont centrés sur l’origine de coordonnées (0,0). Le premier cercle est noir et de rayon 2.
Le pas entre deux rayons est calculé en fonction de $n$ tel que $dr = 2/n$.\
$n$ défini le nombre de cercles à afficher :
- $n = 1$ : 
    - On affiche un seul cercle de rayon 2
- $n = 2$ : 
    - Le premier cercle de rayon 2 est noir
    - Le second cerlce de rayon 1 est blanc
- $n = 3$ : 
    - Le premier cercle de rayon 2 est noir
    - Le second cerlce de rayon 1.34 est blanc
    - Le troisième cerlce de rayon 0.68 est noir

Avec matplotlib, noir c’est `'k'` , et blanc c’est `'w'`.    

````python
import matplotlib.pyplot as plt

# trace un cercle de rayon r et de couleur c
def cercle(r,c):
    x = 0
    y = 0
    circle = plt.Circle((x,y), r, color=c)
    plt.gca().add_artist(circle)

# affiche la figure en cours de taille dans un axe x_min,x_max,y_min,y_max. 
def affiche(x_min, x_max, y_min, y_max):
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
````

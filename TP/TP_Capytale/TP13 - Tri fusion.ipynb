# TP13 - Tri fusion

Nous allons voir aujourd'hui un nouvel algorithme de tri, cette fois-ci plus performant que les tris sélection, insertion et bulle.


!!! info principe
Le tri fusion comporte 2 étapes principales : 
- **Diviser** la liste en 2, trier ces 2 sous-listes (par recursion)
- **Fusionner** les 2 listes triées en une seule liste triée
!!!

!!! abstract Algorithme
On obtient alors l'algotihme diviser pour régner suivant :
1. Cas de base :
    - Si la liste est vide, elle est déjà triée
    - Si la liste n'a qu'un élément, elle est déjà triée.
2. Cas récursif :
    - Diviser : Séparer la liste en deux parties à peu près égales, de tailles ⌈n/2⌉ et ⌊n/2⌋ respectivement.
    - Régner : Trier récursivement les deux parties avec l'algorithme du tri fusion.
    - Combiner : Fusionner les deux listes triées en une seule.
!!!

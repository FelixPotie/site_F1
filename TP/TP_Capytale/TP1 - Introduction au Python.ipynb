!!! info Noms des variables
En Python un nom de variable peut commencer par une lettre ou un underscore `_` et peut ensuite contenir des lettres, chiffres et underscore.

En mathématiques, on a tendance à utiliser des noms de variables le plus court possible : $n$, $k$, ...\
En informatique, c'est l'inverse ! On préfère avoir des noms de variables explicites, pour que le code soit plus clair.

En Python, la convention est d'utiliser la snake_case pour le nom des variables. Cela signifie qu'on ne met pas de majuscules (sauf pour les acronymes), et que les mots sont séparés par des underscore _. Je vous demande de respecter cette convention.

## Types

Les valeurs (et donc les variables associées) ont également des types. Par exemple, la variable `a` définie plus haut contient un nombre entier. Le type correspondant est `int`.

!!! info Les types de base en Python
- `int` pour les entiers de n'importe quelle taille
- `float` pour les nombres "à virgule flottante", ou "flottants". C'est une approximation des nombres réels. En informatique, on note la virgule avec le `.` (et non pas une virgule).
- `str` pour le texte (ou "chaine de caractère" en informatique, "string" en anglais). Les chaines de caractères sont écrites entre guillemets simples, doubles ou triples `'` ou `"` ou `'''`.
- `bool` pour les booléens, qui sont des valeurs qui valent soit `True` (vrai) soit `False` (faux).
- `NoneType` pour `None` (que l'on verra plus tard)
- `list` pour les listes de valeurs (que l'on verra plus tard)
!!!

## Listes

!!! info Le type `list`
Le type `list` permet de stocker plusieurs valeurs ordonnées. En python, on peut voir ça comme un tableau qui n'a pas de fin. (Dans d'autres langages la différence est plus importante).

!!!
!!! note Syntaxe liste
`[]` : liste vide\

`t[i]` : pour accéder à l'élément à l'indice `i` de `t`\
`len(t)` : pour avoir la taille de `t`\
`t.append(x)` : ajoute la valeur de `x` à `t`

## Boucles

Une boucle est un moyen de répéter un bloc de code plusieurs
fois. Il existe 2 méthodes pour faire des boucles, les `for` et les `while`.

!!! note Syntaxe du `for`
La boucle `for` ou "pour" en français, permet d'itérer sur une liste d'éléments et se note :

```python
for variable in valeurs :
    #faire quelque chose

```
Les `valeurs` sont sous forme d'un objet itérable (souvent une liste ou un range). Lors de l'exécution du code, `variable` prend successivement les valeurs de `valeurs` une à une. 
!!!

On utilise fréquemment la fonction `range()` pour coder des boucles.

````python
for i in [0,1,"Bonjour",3.3, True]:
    print(i)
````

````python
for i in range(5):
    print(i)
````

````python
for x in 'Bonjour':
    print(x)
````


!!! note Syntaxe du `while`
La boucle `while` ou "tant que" en français, permet d'exécuter un bloc d'instructions jusqu'à ce qu'une condition nous dise de sortir de la boucle. Cette condition est appelée *la condition d'arrêt*. \
Le while s'écrit :
```python 
while condition :
    #faire quelque chose
#ce qui est écrit ici s'exécutera après la boucle
```
`condition` est un booléen, on sort de la boucle seulement si condition est égale à `False`.

Deux exemples à tester : 
!!!


````python
i = 0
while i < 3:
    print(i)
    i = i + 1       
print("C'est terminé.")
````

````python
i = 0
while i < 3:
    print(i)
    i = i + 1       
    print("C'est terminé.")
````

## Fonctions

Une fonction est un fragment de code réutilisable réalisant une
tâche donnée. Une fonction possède un nom, un nombre quelconque de paramètres et renvoie un résultat ou non.

!!! note Syntaxe de fonction
```python
# implémentation de la fonction
def nom_de_la_fonction(parametre1, parametre2):
    # faire quelque chose
    return ...
#utilisation de la fonction
nom_de_la_fonction(3, "Bonjour")
```

Comme pour les variables, les noms des fonctions sont en minuscule, commencent par une lettre et peuvent contenir des underscores et des chiffres.

Une fonction peut retourner un résultat ou non. Le mot `return` n'est donc pas toujours nécessaire. Une fonction sans return revient à faire une fonction qui se termine par `return None`. `None` est la valeur par défaut, on verra son utilité plus en détail par la suite.

!!!

!!! question Question 16
Testez les fonctions suivantes et comprenez la différence entre une fonction qui retourne un résultat et une qui l'affiche.
!!!

````python
def ajoute_1_sans_return(x):
    x = x + 1
    print("La valeur de x dans la fonction : " + str(x))
    
x = 3
ajoute_1_sans_return(x)
print("La valeur de x en après de la fonction : " + str(x))
# x n'a pas été modifié ! En général il faut toujours retourner un résultat
````

````python
def ajoute_1_avec_return(x):
    return x + 1

x = 3
x = ajoute_1_avec_return(x) # penser à affecter la nouvelle valeur à x
print("La valeur de x en dehors de la boucle : " + str(x))
````

Description formelle et précise de ce que doit faire une fonction. La spécification d'une fonction contient : 
  - des préconditions sur les entrées
  - des postconditions sur les sorties.

On l'écrit en commentaire au début de la fonction.

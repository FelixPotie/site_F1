## Exercice 3 : Arrondi de flottant
!!! info
On reprend l'algorithme glouton du rendu de monnaie. Cette fois-ci on ajoute les centimes. (Vous allez maintenant comprendre pourquoi je les ai volontairement enlever dans le TP14).
!!!

!!! question Question 1
Exécuter les cellules suivantes et expliquer ce qu'il se passe.
!!!

```python
monnaie = [500, 200, 100, 50, 20, 10, 5, 2, 1, 0.5, 0.2, 0.1, 0.05, 0.02, 0.01]
```

```python
def piece_billet_max_1(reste):
    i = 0
    while i < len(monnaie) and reste < monnaie[i]:
        i+= 1
    return monnaie[i]
```

```python
def rendu_monnaie(somme):
    rendu = []
    while somme > 0 :
        print('valeur de la somme à rendre :' + str(somme))
        prochain = piece_billet_max_1(somme)
        rendu.append(prochain)
        somme -= prochain
    return rendu
```

```python
print(rendu_monnaie(123.6))
```

## Exercice 2 : Exponentiation rapide
!!! info
On note $⌊n/2⌋$ la partie entière de $n/2$.\
Soient les 4 fonctions suivantes permettant toutes de calculer $a^n, n \in \mathbb{N}$ :

- `exp_1(a,n)` : $a^n = a^n$

- `exp_2(a,n)` : $a^n = a^{n-1} * a$

- `exp_3(a,n)` : $a^n = a^{⌊n/2⌋} * a^{n-⌊n/2⌋}$

- `exp_rapide(a,n)` réalise un calcul en exploitant : $a^n = \begin{cases}
    (a^{⌊n/2⌋})²       & \quad \text{si } n \text{ est pair}\\
    (a^{⌊n/2⌋})² * a  & \quad \text{sinon} 
  \end{cases}
  $

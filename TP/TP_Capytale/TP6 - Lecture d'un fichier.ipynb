# TP6 - Lecture dans un fichier

L'objectif de ce TP est d'apprendre à extraire des informations depuis un fichier externe.

!!! info Ouverture d'un fichier 
Pour pouvoir lire le contenue d'un fichier, on va d'avord devoir l'ouvrir (comme un livre) puis on pourra le lire ligne par ligne (on lit dans l'ordre), puis il faut bien penser à fermer le fichier (on ferme un livre avant de le ranger).
!!!

!!! abstract open()
`fichier = open(nom_fichier, mode)` : sert donc à ouvrir le fichier `nom_fichier` avec le mode `mode`. Il renvoie un descripteur de fichier qui va nous servir à interragir avec le fichier.

Le mode sert à dire ce qu'on veut faire avec le fichier :
- `'r'` open for reading (default)
- `'w'` open for writing, truncating the file first (écrase ce qu'il y a dans le fichier)
- `'x'` create a new file and open it for writing 
- `'a'` open for writing, appending to the end of the file if it exists

Dans ce TP on va se concentrer sur la lecture de fichier, donc le mode `'r'`.

!!!
!!! info Nom du fichier
Attention le fichier doit se trouver dans le répertoire courant ou alors il faut mettre le chemin vers le fichier complet (ex : \Users\felix\note.txt). \
Dans Capytale c'est un peu différent (car dans le cloud), vous trouverez les fichiers dans `Capytale > Liste des fichiers annexes > attachés à l'activité` (ou en passant par le boutton avec le logo trombonne).\
Vous n'aurez donc qu'à mettre le nom du fichier. 
!!!

!!! abstract fichier.close()
`fichier.close()` : ferme le descripteur de fichier, après avoir fini la lecture. \
Permet de libérer les ressources systèmes. Chaque système d'exploitation n'a qu'un nombre restreint de fichiers pouvant être ouvert en même temps.
!!!
!!! abstract `fichier.readline()` et `fichier.readlines()`
On peut :
- soit lire ligne par ligne un fichier (ouvert en lecture) avec `fichier.readline()`,
- soit lire toutes les lignes et les stocker dans une liste avec `fichier.readlines()`.
!!!

````python
# on souhaite ouvrir le fichier 'Premiers_1000.txt' 
nom_fichier = 'Premiers_1000.txt' 

# on ouvre le fichier en lecture
fichier = open(nom_fichier,'r')

# on récupère toutes les lignes du fichier qu'on stocke dans une variable
lignes = fichier.readlines()

# on ferme le fichier 
fichier.close()
````

La fonction `string.split(char)` permet de séparer un string en liste de string en fonction d'un caractère (si aucun caractère n'est présent c'est l'espace `' '` par défaut.

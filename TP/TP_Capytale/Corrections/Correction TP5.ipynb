!!! question Question 1
Écrivez une fonction `echange(t, i, j)` qui échange les éléments aux indices `i` et `j` de la liste `t`. 
!!!

```python
def echange(t, i, j):
    t[i], t[j] =  t[j], t[i]
```

```python
l = [1,2,3,4]
echange(l, 1, 3)
print(l) # [1, 4, 3, 2]
```

### Premier algorithme

Soit `t` une liste de `n` entiers.

- On parcourt la liste en inversant `t[i]` et `t[i+1]` chaque fois que `t[i] > t[i+1]`.
- On recommence tant qu'il y a eu des échanges au parcours précédent.

!!! question Question 2
Créer une fonction `parcours_1(t)` qui parcours une fois `t`, fait les échanges si besoin et renvoie `True` ssi au moins un échange a été réalisé.\
Comment est `t` si `parcours_1(t)` renvoie faux ?
!!!

```python
def parcours_1(t):
    changement = False
    for i in range(len(t)-1):
        if t[i] > t[i+1]:
            echange(t, i, i+1)
            changement = True
    return changement
```

```python
# Test
t = [7,6,2,3,8]
res = parcours_1(t) 
print(res) # True
print(t) # [6,2,3,7,8]
```

```python
# Test
t = [2,3,6,7,8]
res = parcours_1(t) 
print(res) # False
print(t) # [2,3,6,7,8]
```

!!! question Question 3
Créer la fonction `tri_bulle_1(t)` qui tri `t` en utilisant la fonction `parcours_1(t)`.
!!!

```python
def tri_bulle_1(t):
    changement = True
    while changement:
        changement = parcours_1(t)
```

### Second algorithme

- On parcourt la liste `n-1` fois en inversant `t[i]` et `t[i+1]` chaque fois que `t[i] > t[i+1]`.
- La k-ème fois, on ne touche pas au k-1 derniers éléments (déjà triés)

!!! question Question 5
Créer une fonction `parcours_2(t,k)` avec `k>=0` sur la base de `parcours_1` parcourant la liste `t` sans toucher à ses `k-1` derniers éléments en procédant aux inversions vues précédemment.
!!!

```python
def parcours_2(t,k):
    assert k >= 0, "k doit être positif"
    
    changement = False
    for i in range(len(t)-k-1):
        if t[i] > t[i+1]:
            echange(t, i, i+1)
            changement = True
    return changement
```

```python
# Test
t = [7,6,2,3,8]
res = parcours_2(t,1) 
print(res) # True
print(t) # [6,2,3,7,8]
res = parcours_2(t,2) 
print(res) # True
print(t) # [2,3,6,7,8]
res = parcours_2(t,3) 
print(res) # False
```

!!! question Question 6
Créer la fonction `tri_bulle_2(t)` qui réalise ce tri.
!!!

```python
def tri_bulle_2(t):
    changement = True
    k = 0
    while changement:
        changement = parcours_2(t,k)
        k += 1
```

```python
t = [7,6,2,3,8]
tri_bulle_2(t)
print(t)
```

!!! question Question 9
Donner les compléxités dans le pire cas, en grand O, de vos 2 fonctions de tri bulle.
!!!

On pose $n$ la taille de la liste `t`\
**tri_bulle_1:**\
`parcours_1` à une complexité de $O(n)$, dans le pire des cas (liste trié par ordre décroissant) `tri_bulle_1` appelle $n$ `parcours_1`. On a alors une complexité en $O(n*n)=O(n²)$

**tri_bulle_2:**\
`parcours_2` à une complexité de $O(n-k)$, dans le pire des cas (liste trié par ordre décroissant) `tri_bulle_2` appelle $n$ `parcours_2` avec un $k$ de plus en plus grand.\
On a donc $\sum_{k=0}^n n-k$ = $n(n+1) \over 2$ $= O(n²)$ \
On a alors une complexité en $O(n²)$ dans les 2 cas, même si la 2eme version est bien plus performante.

## Exercice Bonus : Tri cocktail

Un dérivé du tri à bulles est le tri *cocktail* ou tri *shaker*. Cette méthode de tri est basée sur l'observation suivante : dans le tri à bulles, les éléments peuvent avancer rapidement vers la fin du tableau, mais ne sont déplacés vers le début du tableau que d'une position à la fois.

L'idée du tri cocktail consiste à alterner le sens du parcours.
!!! question Question 1
Implémenter la fonction **tri_cocktail(t)**.
!!!


```python
def parcours_3(t, k, forward):
    assert k >= 0, "k doit être positif"
    
    changement = False
    if forward : 
        for i in range(k, len(t)-k-1):
            if t[i] > t[i+1]:
                echange(t, i, i+1)
                changement = True
    else : #backword
        for i in range(len(t)-k-1, k-1, -1):
            if t[i] < t[i-1]:
                echange(t, i, i-1)
                changement = True
    return changement
```

```python
def tri_cocktail(t):
    changement = True
    forward = True
    k = 0
    while changement:
        changement = parcours_3(t, k, forward)
        if forward:
            k += 1
        forward = not forward
```

```python
t = [7,6,2,3,8,1]
tri_cocktail(t)
print(t)
```

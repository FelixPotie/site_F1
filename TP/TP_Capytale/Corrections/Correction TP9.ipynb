# TP9 - Tri comptage

!!! info Tri par comparaison
On a vu les algorithmes de tris suivant : tri par selection, tri par insertion et tri bulle. Ces tris sont des tris appelés tri par comparaison. Ils utilisent un enchainement de comparaisons (`<` ou`>`), cependant il existe d'autres type de tris qui ne sont pas par comparaison. Nous allons en voir un aujourd'hui.
!!!

!!! note Version 1
Pour commencer, nous allons faire une version où l'on veut trier des entiers naturels compris dans une borne `[0:m]`, où `m` est l'élément maximum de la liste. 

Le principe du tri par comptage est de :
1. Compter le nombre d'occurence de chaque chiffre de la liste, on le stocke dans une liste `compteurs` de taille `m+1` où `compteurs[i]` contiendra le nombre d'occurence de `i` dans la liste.
2. Initialiser une liste `liste_triee` vide 
2. Parcourir chaque éléments `e` de `compteurs` et ajouter dans `liste_triee` le nombre `e` le nombre de fois qu'il faut (`compteurs[i]`).

Par exemple si on a la liste : \
`liste = [1,3,2,2,1]`\
alors va créer :\
`compteurs = [0,2,2,1]`\
et donc on aura :\
`liste_triee = [1,1,2,2,3]`.
!!!

!!! question Question 1 
Implémentez une fonction `tri_comptage_1(liste)` à l'aide d'un tableau et de la fonction `max()`.
!!!

````python
def tri_comptage_1(liste):
    # on initialise notre liste de compteurs
    compteurs = [0 for i in range(max(liste)+1)]
    
    # on compte le nombre d'occurence de chaque entier entre  [0:max(liste)]
    for element in liste:
        compteurs[element] += 1
        
    # on vient reconstituer notre liste triée
    liste_triee = []
    for i in range(len(compteurs)): 
        for _ in range(compteurs[i]):
            liste_triee.append(i)
    return liste_triee
````

````python
print(tri_comptage_1([1,3,2,2,1])) #[1,1,2,2,3]
````

3. Dans quel cas cette fonction n'a pas une bonne compléxité ?

**1**\
On note $n$ la taille de la liste et $m$ l'element maximum de liste.
max(liste) : $O(n)$
init de compteurs : $O(m)$
complétion de compteurs : $O(n)$
remplissage de liste_triee : Attention, ici on a $O(n)$, en effet on a 2 boucles for, mais la pluspart d'entre elle ne font pas d'opérations, au total on a bien $n$ éléments à mettre dans notre liste.

On a donc $O(n+m+n+n) = O(n+m)$

**2**\
Lorsque $m$ est inférieur à $n²$.

**3**\
Cette fonction à une mauvaise compléxité si les données sont très éloignées et disparate.

!!! question Question 3
On va ré-écrire l'algorithme du tri par insertion en le décomposant.\
Écrivez une fonction `indice_ins(liste : list[int], e : [int]) -> int` qui prend une liste `liste` triée et renvoie l'indice auquel insérer `e` pour que la liste reste triée. 
!!!


````python
def indice_ins(liste, e):
    i = 0
    while i < len(liste) and liste[i] < e:
        i += 1
    return i
````

````python
print(indice_ins([1,3,4], 2)) #1
print(indice_ins([1,3,4], 0)) #0
print(indice_ins([1,3,4], 6)) #3
````

Écrire une fonction `insert(liste, e, i)` qui insert e dans liste à l'indice i grâce à des **slices**.
!!!

````python
def insert(liste, i, e):
    return liste[:i] + [e] + liste[i:]
````

!!! question Question 5
Implémenter maintenant le tri par insertion avec les fonction `indice_ins(liste, e)` et la fonction d'insertion de votre choix.
!!!

````python
def tri_insertion(liste):
    res = []
    for e in liste:
        i = indice_ins(res, e)
        res.insert(i, e)
    return res
````

````python
print(tri_insertion([5,3,6,2,9,8,1,7,4,0]))
````

!!! question Question 6
Implémentez une fonction `tri_comptage_2(liste)` à l'aide d'un dictionnaire et de votre fonction de tri de la question précédente.

1. Initialisation d'un dictionnaire vide
2. On parcourt la liste :
    - si l'élément est présent dans le dico on incrémente sa valeur
    - sinon on l'ajoute dans le dico avec une valeur de 1
3. Récupérer les clés du dictionnaires, les transformer en liste et les triés grace à votre fonction de tri de la Q3
4. Reconstuire la liste finale triée
!!!

````python
def tri_comptage_2(liste):
    dico = {}
    for element in liste:
        if element in dico:
            dico[element] += 1
        else:
            dico[element] = 1
    cles = list(dico.keys())
    cles = tri_insertion(cles)
    liste_triee = []
    for cle in cles:
        for _ in range(dico[cle]):
            liste_triee.append(cle)
    return liste_triee
````

````python
print(tri_comptage_2([74230,75012,34200,38400,8000]))
````

!!! question Question 7
On va maintenant améliorer notre fonction de recherche d'indice `indice_ins(liste, e)` en faisant cette recherche d'indice avec le recherche dichotomique.

Écrivez la fonction `indice_ins_dicho(liste, e)`.
!!!

````python
def indice_ins_dicho(liste, e):
    debut = 0
    fin = len(liste)-1
    while debut <= fin:
        milieu = (fin + debut) // 2
        if e == liste[milieu]:
            return milieu
        elif e < liste[milieu]:
            fin = milieu - 1
        else:
            debut = milieu + 1
    return debut
````

````python
#autre version proche
def indice_ins_dicho(liste, e):
    debut = 0
    fin = len(liste)
    while debut < fin:
        milieu = (debut+fin) // 2
        if liste[milieu] < e:
            debut = milieu + 1
        else:
            fin = milieu
    return fin #debut et fin sont égaux
````

````python
print(indice_ins_dicho([1,3,4], 2)) #1
print(indice_ins_dicho([1,3,4], 0)) #0
print(indice_ins_dicho([1,3,4], 6)) #3
````

!!! question Question 8
On va encore améliorer les performance de notre algorithme, pour cela on va utiliser notre fonction `indice_ins_dicho` et on va utiliser la propriété **online** du tri par insertion : 

1. Initialisation d'un dictionnaire vide et d'une liste de clé vide
2. On parcourt la liste :
    - si l'élément est présent dans le dico on incrémente sa valeur
    - sinon :
        - on l'ajoute dans le dico avec une valeur de 1
        - on l'ajoute dans notre liste de clé au bon endroit pour que notre liste reste trié (grace à `indice_ins_dicho(l,e)`)
3. Reconstuire la liste finale triée

Implémentez cette fonction qu'on appellera `tri_comptage_3(liste)`.
!!!

````python
def tri_comptage_3(liste):
    dico = {}
    cles = []
    for element in liste:
        if element in dico:
            dico[element] += 1
        else:
            dico[element] = 1
            i = indice_ins_dicho(cles, element)
            cles.insert(i, element)
    liste_triee = []
    for cle in cles:
        for _ in range(dico[cle]):
            liste_triee.append(cle)
    return liste_triee
````

À l'aide d'un dictionnaire trouvé quel code postal contient le plus d'associations RUP.
!!!

````python
occ = {}
for code in codes_postaux:
    if code in occ:
        occ[code] += 1
    else :
        occ[code] = 1 
code_max = None
for code, nb in occ.items():
    if code_max is None or occ[code_max] < nb:
        code_max = code
print(f"Le code postal {code_max}, a le maximum d'assos RUP avec {occ[code_max]} assos")
````

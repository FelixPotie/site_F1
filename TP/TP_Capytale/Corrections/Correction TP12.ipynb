# TP 12 - Enumération des sous listes ou des permutations d’une liste
!!! info 
Soit une liste que nous limiterons pour le moment à des entiers. On souhaite mettre en place un
algorithme récursif qui permette de lister toutes les sous-listes de `l` avec `k` éléments. 

Soit une liste `l=[1,2,3,4]` et `k=2`.
Nous attendons deux types de résultats :
- V1 (enum_t) : Les sous-listes avec toutes les permutations possibles :\
`[[1,2],[1,3],[1,4],[2,1],[2,3],[2,4],[3,1],[3,2],[3,4],[4,1],[4,2],[4,3]]`
- V2 (enum_u) : Les sous-listes sans les permutations :\
`[[1,2],[1,3],[1,4],[2,3],[2,4],[3,4]]`
!!!
!!! abstract 
Le principe de l’algorithme est le suivant :
- traiter le cas de base lorsque `k=1`
- créer par récursivité l’énumération de plusieurs sous listes de `l` de `k−1` éléments et de leur ajouter celui qui leur manque (je ne dis volontairement pas tout). On ajoute alors tous les résultats attendus en `k` éléments dans une liste qui est renvoyée.
!!!
!!! example Algo
- Cas `k` = 1 : 
    - On retourne la liste de liste des éléments de `l`
- Sinon :
    - On crée une liste résultat `res` vide
    - Pour `e` chaque éléments de `l`:
        - on récupère une sous-liste de `l` sans `e`
        - on récupère le résultat de `enum_t` sur notre sous-liste avec `k-1`
        - Pour chaque liste de ce résultat (liste) :
            - On ajoute dans `res` la liste dans laquelle on a ajouté `e`
    - On retourne `res`
!!!
!!! question Question 1
Créer une fonction récursive `enum_t(l,k)` qui renvoie les sous-listes de type V1 de `l` de `k` éléments.

**Prenez un brouillon pour comprendre comment doit fonctionner votre algorithme !**

!!!



```python
def enum_t(l,k):
    if k == 1 :
        return [[e] for e in l]
    res = []
    for i in range(len(l)):
        # on récupère toutes les sous listes sans l[i] 
        sous_listes_1 = enum_t(l[:i]+l[i+1:],k-1)
        # on rajoute l[i] sur chacune de ces sous-listes
        for s_l in sous_listes_1:
            res.append([l[i]]+s_l)
    return res
```

!!! question Question 2
Créer une fonction récursive `enum_u(l,k)` qui renvoie les sous-listes de type V2 de `l` de `k` éléments.

_indice : on va récupérer les sous-listes de `k-1` eléments sans les `i` premiers éléments de `l`, donc par exemple les sous-liste de taille 1 de `[2,3,4]` puis rajouter `1` à chacune de ces sous-liste on aura donc `[[1,2],[1,3],[1,4]` et il faudra ensuite faire pareil en enlevant les 2 premiers éléments puis les 3..._

```python
def enum_u(l,k):
    if k == 1 :
        return [[e] for e in l]
    res = []
    for i in range(len(l)):
        sous_listes_1 = enum_u(l[i+1:],k-1)
        for s_l in sous_listes_1:
            res.append([l[i]]+s_l)
    return res
```

### Application 2 : Scrabble
!!! question Question 6
Adapter votre fonction `enum_t` en une fonction `enum_t_str(texte, k)` capable de traiter une chaine de caractères de type `str`.
!!!

```python
def enum_t_str(l,k):
    if k == 1 :
        return [e for e in l]
    res = []
    for i in range(len(l)):
        # on récupère toutes les sous listes sans l[i] 
        sous_listes_1 = enum_t_str(l[:i]+l[i+1:],k-1)
        # on rajoute l[i] sur chacune de ces sous-listes
        for s_l in sous_listes_1:
            res.append(l[i]+s_l)
    return res
```

```python
# ['uo', 'uc', 'ou', 'oc', 'cu', 'co']
# ou 
# ['ou', 'cu', 'uo', 'co', 'uc', 'oc']
```

```python
print(enum_t_str('uoc', 3))
# ['uoc', 'uco', 'ouc', 'ocu', 'cuo', 'cou']
# ou 
# ['cou', 'ocu', 'cuo', 'uco', 'ouc', 'uoc']
```

!!! question Question 7
En vous aidant de votre fonction `enum_t_str`, trouver le plus long mot que l’on peut écrire avec les lettres "vrboa". Vous disposez d'un fichier `liste_francais.txt` qui contient une liste de mot du dictionnaire français.
!!!

```python
fichier = open("liste_francais.txt", 'r')
lignes = fichier.readlines()
mots = [m.strip() for m in lignes]
fichier.close()

texte = 'vrboa'
for e in enum_t_str(texte, 5):
    if e in mots : 
        print(e)
```

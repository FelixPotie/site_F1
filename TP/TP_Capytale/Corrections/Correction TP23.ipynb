!!! question Question 1
Dans un premier temps on va configurer notre grille de la façon suivante : 
- matrice de taille $n*n$ (choix d'une matrice carrée par simplicité)
- une case de la matrice vaut :
    - 0 : si elle est vide 
    - 1 : si un mur/obstacle est présent
    - 2 : si c'est le point de départ (un seul)
    - 3 : si c'est le point de sortie (un seul)
    - 4 : (pour la visualisation de Dijkstra et A*) si elle est vide et qu'elle est dans la file de priorité
    - 5 : (pour la visualisation de Dijkstra et A*) si elle est vide et qu'elle est sortie de la file de priorité 
    
Écrire une fonction `creer_grille(n: int, depart: tuple, arrivee: tuple) -> list[list[int]]` qui créé une grille comme décrit ci-dessus, pour l'instant sans mur. `depart` et `arrivee` sont des tuples qui représentent les coordonnées des points.
!!!

````python
def creer_grille(n, depart, arrivee):
    grille = [[0 for _ in range(n)] for _ in range(n)]
    grille[depart[0]][depart[1]] = 2
    grille[arrivee[0]][arrivee[1]] = 3
    return grille
````

!!! info
La fonction suivante permet d'afficher la grille créée par votre fonction

````python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap


def afficher_grille(grille, chemin = None):
    """
    affiche la grille avec les murs et les cases visitées
    un paramètre optionnel trace le chemin
    """
    # Création d'une palette de couleurs personnalisée
    # 0: Blanc, 1: Gris foncé (Mur), 2: Vert (Start), 3: Rouge (Target), 4: Jaune (Dans file prio), 5: Bleu (Visité)
    cmap = ListedColormap(['white', '#404040', '#00D100', '#e74c3c', '#FAFF69', '#B8BCFF' ])

    fig, ax = plt.subplots()
    
    # Affichage de la grille 
    ax.imshow(grille, cmap=cmap, vmin=0, vmax=5)
    
    if chemin and len(chemin) > 0:
        # On sépare les lignes (r) et les colonnes (c)
        # Attention : r correspond à l'axe Y et c à l'axe X dans imshow
        y, x = zip(*chemin)
        
        ax.plot(x, y, color='#e74c3c', linewidth=3, zorder=5)

    # On cache les chiffres des axes pour un look plus "jeu"
    ax.tick_params(which='both', bottom=False, left=False, labelbottom=False, labelleft=False)

    plt.show()
````

````python
n=30
depart = (15, 5)
arrivee = (15, 25)
grille = creer_grille(n,depart, arrivee)
afficher_grille(grille)
````

Nous n'allons pas créer un vrai graphe par dictionnaire d'adjacence. Pour représenter les déplacements possibles nous allons juste coder une fonction qui permet de récupérer les successeurs d'une case dans notre grille.

Écrire une fonction `voisins(grille, x: int, y: int)-> list[tuple]` qui renvoie la liste des voisins (cases) accessibles d'un point aux coordonnées (x,y). On part du principe qu'on ne peut pas se déplacer en diagonale. Attention, « accessible » signifie que la valeur est égale à 0 ou 3 (vide ou case arrivee).\
Attention à ne pas sortir de la grille !

Depuis la case bleue on peut aller sur les cases vertes.
<img src = "https://capytale2.ac-paris.fr/web/sites/default/files/2026/04-08/12-00-46/Capture%20d%E2%80%99%C3%A9cran%20du%202026-04-08%2011-57-08.png" style="width: 30%" />
!!!


````python
def ajoute_si_libre(g, k, l, res):
    if g[k][l] == 0 or g[k][l] == 3 :
        res.append((k,l))

def voisins(g, x, y):
    res = []
    if x != 0:
        ajoute_si_libre(g, x-1, y, res)
    if y != 0:
        ajoute_si_libre(g, x, y-1, res)
    if x != len(g)-1:
        ajoute_si_libre(g, x+1, y, res)
    if y != len(g)-1:
        ajoute_si_libre(g, x, y+1, res)
    return res
                    
````

````python
import heapq
import copy

def dijkstra(grille, depart, arrivee):
    # initialisation nouvelle grille et file de prio
    g = copy.deepcopy(grille)
    fp = [(0, depart)]
    
    while len(fp) != 0:
        # on défile l'élément le plus petit
        dist_u, u = heapq.heappop(fp)
        # on passe la case en bleu si jaune
        if g[u[0]][u[1]] == 4:
            g[u[0]][u[1]] = 5

        for v in voisins(g, u[0], u[1]):
            if v == arrivee :
                return g
            else :
                heapq.heappush(fp, (dist_u + 1, v))
                # si blanc, on passe en jaune
                if g[v[0]][v[1]] == 0: 
                    g[v[0]][v[1]] = 4 
    return g
````

!!! question Question 4
Écrire une fonction `dm(a, b) -> int` qui calcule la distance de Manhattan : 
$dm(a, b) = |x_a - x_b| + |y_a - y_b|$
!!!

````python
def dm(a,b):
    return abs(a[0]-b[0])+abs(a[1]-b[1])
````

!!! question Question 5
Écrire une fonction `a_star(grille, depart, arrivee)-> list[list]]` qui implémente l'algorithme A* avec l'heuristique totale distance de Manhattan + distance depuis la source.

_Vous repartirez de votre code de Dijkstra et ferez les modifications nécessaires._

_Indice : la file de priorité contiendra maintenant des 3-uplet (heuristique totale, dist, sommet)_
!!!

````python
import heapq
import copy

def a_star(grille, depart, arrivee):
    # initialisation nouvelle grille et file de prio
    g = copy.deepcopy(grille)
    fp = [(dm(depart, arrivee), 0, depart)]
    
    while len(fp) != 0:
        # on défile l'élément le plus petit
        ht_u, dist_u, u = heapq.heappop(fp)
        # on passe la case en bleu si jaune
        if g[u[0]][u[1]] == 4:
            g[u[0]][u[1]] = 5

        for v in voisins(g, u[0], u[1]):
            if v == arrivee :
                return g
            else :
                dist_v = dist_u + 1
                ht_v = dm(v, arrivee) + dist_v
                heapq.heappush(fp, (ht_v, dist_v, v))
                # si blanc, on passe en jaune
                if g[v[0]][v[1]] == 0: 
                    g[v[0]][v[1]] = 4 
    return g
````

````python
n=30
depart = (15, 5)
arrivee = (15, 25)
grille = creer_grille(n,depart, arrivee)

g_d = dijkstra(grille, depart, arrivee)
g_a = a_star(grille, depart, arrivee)
afficher_grille(grille)
afficher_grille(g_d)
afficher_grille(g_a)
````

!!! question Question 6
Tester vos fonctions sur des labyrinthes générés aléatoirement avec le code suivant et regarder si A* trouve un chemin en explorant moins de cases que dijkstra. _(Vous n'avez pas besoin de comprendre le code suivant qui génère un labyrinthe.)_
!!!

````python
import random

def generer_labyrinthe(n, depart, arrivee, porosite=0.15):
    # 1. Initialisation : Tout est un mur
    grille = [[1 for _ in range(n)] for _ in range(n)]
    
    def est_valide(r, c):
        return 0 <= r < n and 0 <= c < n

    # 2. Algorithme de base 
    stack = [depart]
    grille[depart[0]][depart[1]] = 0
    
    while stack:
        curr_r, curr_c = stack[-1]
        voisins = []
        for dr, dc in [(0, 2), (0, -2), (2, 0), (-2, 0)]:
            nr, nc = curr_r + dr, curr_c + dc
            if est_valide(nr, nc) and grille[nr][nc] == 1:
                voisins.append((nr, nc, curr_r + dr//2, curr_c + dc//2))
        
        if voisins:
            next_r, next_c, wall_r, wall_c = random.choice(voisins)
            grille[wall_r][wall_c] = 0
            grille[next_r][next_c] = 0
            stack.append((next_r, next_c))
        else:
            stack.pop()

    # 3. AJOUT DE CHEMINS MULTIPLES (Porosité)
    # On parcourt les murs intérieurs et on en casse certains au hasard
    for r in range(1, n - 1):
        for c in range(1, n - 1):
            if grille[r][c] == 1:
                # Si le mur sépare deux couloirs horizontaux ou verticaux
                # On a une chance de le casser pour créer une boucle
                if random.random() < porosite:
                    grille[r][c] = 0

    # 4. Placement final du depart et de l'arrivée (3)
    # On s'assure que l'arrivée n'est pas un mur
    grille[arrivee[0]][arrivee[1]] = 0 
    
    # Si l'arrivée est isolée (entourée de murs), on force une ouverture
    if all(grille[arrivee[0]+dr][arrivee[1]+dc] == 1 
           for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)] 
           if est_valide(arrivee[0]+dr, arrivee[1]+dc)):
        # On casse un mur adjacent au hasard
        grille[arrivee[0]-1][arrivee[1]] = 0

    grille[depart[0]][depart[1]] = 2
    grille[arrivee[0]][arrivee[1]] = 3

    return grille
````

!!! question Question 7
Reprenez vos fonctions `dijkstra` et `a_star` pour retourner également un dictionnaire `parents` qui associe une case à la case depuis laquelle on est arrivé, pour permettre de reconstruire le chemin.
!!!

````python
import heapq
import copy

def dijkstra(grille, depart, arrivee):
    # initialisation nouvelle grille et file de prio
    g = copy.deepcopy(grille)
    fp = [(0, depart)]
    parents = {depart: None}
    
    while len(fp) != 0:
        # on défile l'élément le plus petit
        dist_u, u = heapq.heappop(fp)
        # on passe la case en bleu si jaune
        if g[u[0]][u[1]] == 4:
            g[u[0]][u[1]] = 5

        for v in voisins(g, u[0], u[1]):
            parents[v]=u
            if v == arrivee :
                return g, parents
            else :
                heapq.heappush(fp, (dist_u + 1, v))
                # si blanc, on passe en jaune
                if g[v[0]][v[1]] == 0: 
                    g[v[0]][v[1]] = 4 
````

````python
import heapq
import copy

def a_star(grille, depart, arrivee):
    # initialisation nouvelle grille et file de prio
    g = copy.deepcopy(grille)
    fp = [(dm(depart, arrivee), 0, depart)]
    parents = {depart: None}
    
    while len(fp) != 0:
        # on défile l'élément le plus petit
        ht_u, dist_u, u = heapq.heappop(fp)
        # on passe la case en bleu si jaune
        if g[u[0]][u[1]] == 4:
            g[u[0]][u[1]] = 5

        for v in voisins(g, u[0], u[1]):
            parents[v]=u
            if v == arrivee :
                return g, parents
            else :
                dist_v = dist_u + 1
                ht_v = dm(v, arrivee) + dist_v
                heapq.heappush(fp, (ht_v, dist_v, v))
                # si blanc, on passe en jaune
                if g[v[0]][v[1]] == 0: 
                    g[v[0]][v[1]] = 4 
````

!!! question Question 8
Écrire une fonction `reconstruire_chemin(parents, dest)` qui reconstruit le chemin depuis le dictionnaire parents de la destination jusqu'à la source.
!!!

````python
def reconstruire_chemin(parents, dest):
    chemin = []
    while dest != None:
        chemin.append(dest)
        dest = parents[dest]
    return chemin[::-1]
````

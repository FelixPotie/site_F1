!!! abstract Bibiliothèque
Une bibliothèque est une collection de modules, qui contiennent des fonctions déjà coder que l'on peut réutiliser sans avoir à les reprogrammer.
En Python, il existe de nombreuses bibliothèque, nous allons en voir quelques unes dans l'année.
!!!

!!! abstract Import de module
Pour utiliser une bibliothèque il faut l'importer et donc ajouter une ligne au début du programme qui dit laquelle on souhaite utiliser. Il existe plusieurs manière d'importer un module d'une bibliothèque :
```python
import nom_module # Le module entier est importé. Pour appeler ses éléments il faut faire nom_module.nom_fonction

from nom_du_module import element1, element2 # Seulement element1 et element2 sont importés on peut directement les utiliser 

from nom_du_module import * # tous les élements du module sont importés, pas besoin de remettre le nom du module pour l'utiliser

import nom_module as alias # Import tout le module, pour appeler ses éléments il faut faire alias.nom_fonction
```
!!!

Exemple avec le module `math` et la fonction `sqrt(x)` qui renvoie la racine carrée de `x`.

````python
import math
print(math.sqrt(42))
````

````python
from math import sqrt
print(sqrt(42))
````

````python
import math as m
print(m.sqrt(42))
````

!!! info Matplotlib
Pour les représentations graphiques, la bibliothèque matplotlib est la plus connue et facile d'utilisation. On va voir ensemble quelques exemples d'utilisations, vous pourrez trouver toute la documentation sur https://matplotlib.org/.
!!!
!!! abstract pyplot
Pour représenter des données graphiquement on va utiliser le module pyplot de la biblipthèque matplotlib. En général on renomme ce module plt de cette façon :
```python 
import matplotlib.pyplot as plt
```

On initialise la figure (on peut ajouter la taille de la figure avec le paramètre `figsize=(width,height)` mais dans notre notebook nous n'en avons pas besoin) :

```python 
plt.figure(nom)
```

Ensuite, on va dire tout ce qu'on veut afficher (nuages de points, courbes, légendes...).
On peut par exemple faire : 
```python 
plt.plot(x,y)
```

Puis il faut terminer par afficher notre graphique en faisant :
```python 
plt.show()
```
!!!

!!! question Question 0
Regardez l'exemple si dessous, essayez de bouger certains paramètres pour comprendre le fonctionnement de chaque fonction.
!!!

````python
import matplotlib.pyplot as plt

x = [1,2,3,4] # abscisses
y = [1,-5,3,2] # ordonnées
# remarque : x et y doivent avoir la même taille

plt.figure() # initialise la figure on peut mettre un nom à la figure si on veut
plt.title("Mon titre") # titre
plt.xlabel("axe x") # nom axe abscisse
plt.ylabel("axe y") # nom axe ordonnée

plt.plot(x,y, "r+--") # on trace les points aux abscisses x et ordonnées y
# le 3eme paramètre sert à désigner :
#    la couleur ("r"/"g"/"b"/"c"/"y"/"m"/"k") 
#    la forme des points ("o"/"s"/"*"/...) 
#    les traits pour relier (""/"-"/"--"/":")

plt.show() # on affiche la figure
````

!!! question Question 2
On souhaite définir une liste d’abscisses `liste_x` régulièrement espacées avec un nombre de points `n` défini :



Écrire une fonction `liste_x(xmin, xmax, n)` renvoyant la liste_x attendue.
!!!

````python
def liste_x(xmin,xmax,n):
    l = xmax - xmin
    dx = l/(n-1)
    return [xmin+i*dx for i in range(n)]
````

````python
print(liste_x(10,15,11))
# [10.0, 10.5, 11.0, 11.5, 12.0, 12.5, 13.0, 13.5, 14.0, 14.5, 15.0]

print(liste_x(0.5,1,6))
# [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
````

!!! question Question 4
On cherche maintenant à afficher les courbes des fonctions $x \mapsto x^2$ et $x \mapsto \sqrt{x}$ pour $x \in [0;2]$.

Écrire les fonctions `carree(x)` et `racice_carree(x)` qui renvoie respectivement $x^2$ et $\sqrt{x}$.\
Rappel : le module math comporte une fonction qui fait la racine carrée.
!!!

````python
from math import sqrt

def carre(x): 
    return x**2


def racine_carree(x): 
    return sqrt(x)
````

## Exercice 2 : Tracer depuis les données d'un fichier

!!! info 
Vous avez à votre disposition le fichier __"Positions carres.txt"__. Chaque ligne contient 4 valeurs : Temps, x, y, z. Après mesure des accélérations avec un téléphone lors d’un mouvement, celles-ci ont été importées sous Python, post-traitées afin de supprimer les valeurs moyennes puis doublement intégrées afin d’obtenir les coordonnées x, y et z du téléphone au cours du temps.
!!!
!!! question Question 1
Créer les 4 listes de flottants temps, x, y, z avec les colonnes du fichier.
!!!

````python
fichier = open('Positions_carres.txt')
fichier.readline() # Sans la première ligne
lignes_donnees = fichier.readlines()
fichier.close()

temps, x, y, z = [], [], [], []
for ligne in lignes_donnees:
    donnes = ligne.split(' ')
    temps.append(float(donnes[0]))
    x.append(float(donnes[1]))
    y.append(float(donnes[2]))
    z.append(float(donnes[3]))

````

!!! question Question 2
Tracer sur une figures `x` en fonction du temps et `y` en fonction du temps.
!!!

````python
import matplotlib.pyplot as plt
    
plt.figure()
plt.plot(temps,x, "r")
plt.plot(temps,y, "g")

plt.legend(['x en fonction du temps', 'y en fonction du temps'])
plt.show()
````

## Exercice 3 : histogramme

On cherche à vérifier la loi de Benford (https://fr.wikipedia.org/wiki/Loi_de_Benford). Celle-ci prévoit
la proportion de nombre commençant par chaque chiffre si on prend des nombres une séries
statistique issue du monde réel.

Vous avez à votre disposition le fichier __"Balance_commune_2018.csv"__. Ce fichier est un fichier CSV (comma separated values). Chaque ligne du fichier contient 4 valeurs qui sont séparées par des virgules, et sont dans l'ordre suivant :
- Le nom de la commune (attention, certains contiennent une virgule)
- Le code INSEE (du département et de la commune)
- Le numéro du compte
- Le solde du compte (qui peut être positif ou négatif)

!!! question Question 1
A partir du fichier, créer la liste `liste_chiffre` contenant le premier chiffre quand il est différent de 0 de chaque solde.
!!!

````python
fichier = open('Balance_commune_2018.csv')
fichier.readline() # Sans la première ligne
lignes_donnees = fichier.readlines()
fichier.close()

liste_chiffre = []
for ligne in lignes_donnees:
    donnes = ligne.split(',')
    solde = donnes[-1]
    if solde[0] == '-':
        premier_chiffre = int(solde[1])
    else : 
        premier_chiffre = int(solde[0])
    if premier_chiffre != 0:
        liste_chiffre.append(premier_chiffre)
````

!!! question Question 2
Tracer l'histogramme de la série statistique obtenue. Vérifier l'allure attendue par la loi de Benford.

Pour tracer un histograme on utilise `plt.hist(liste_donnees, bins=nb_colonne)`. (On doit toujours initialiser la figure et l'afficher.)
!!!

````python
import matplotlib.pyplot as plt
plt.figure()
plt.hist(liste_chiffre, bins=9)
plt.show()
````

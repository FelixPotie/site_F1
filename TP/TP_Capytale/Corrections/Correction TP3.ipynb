## Exercice 1 : Calcul de sommes doubles

!!! question Question 1 
Ecrire une fonction `somme_double_1(n)` qui, étant donné un entier `n`, renvoie la valeur de la somme 
$$
\sum_{i=1}^n \sum_{j=1}^n i^j
$$
Rappel : $x^y$ s'écrit `x**y`
!!!

```python
def somme_double_1(n):
    res = 0
    for i in range(1,n+1):
        for j in range(1, n+1):
            res += i ** j
    return res
```

```python
print(somme_double_1(5)) #5699
```

Ecrire une fonction `somme_double_2(n)` qui, étant donné un entier n, renvoie les valeurs des deux sommes :
$$ S_1 = \sum_{i=1}^n \sum_{j=i}^n i/j$$ 
$$S_2 = \sum_{j=1}^n \sum_{i=1}^j i/j$$
!!!
!!! info Tuple
En python on peut créer des **tuples** en faisant : 
`(x, y)`
!!!

```python
# Exemple de tuple
def je_retourne_2(x,y):
    return (x,y)

print(je_retourne_2(4,6))
```

```python
def somme_double_2(n):
    s_1 = 0
    s_2 = 0
    for i in range(1, n+1):
        for j in range(i, n+1):
            s_1 += i / j
    for j in range(1, n+1):
        for i in range(1, j+1):
            s_2 += i / j
    return s_1, s_2
```

## Exercice 2 : Addition et multiplication de matrice

!!! info Matrice 
En python on peut faire des matrices, ce sont des liste de liste.
!!!
!!! example Exemple de matrice
On peut initialiser la matrice : 
$$ m =
\begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{pmatrix}
$$
en faisant : 
`m = [[1,2,3], [4,5,6]]`

On peut accéder à la valeur à la ligne `i` et la colonne `j` en faisant `m[i][j]`. Attention les indices commencent toujours à `0`.
!!!

!!! info Taille de matrice
`len(m)` renvoie le nombre de ligne\
`len(m[i])` renvoie le nombre de colonne de la ligne `i`
!!!

```python
m = [[1,2,3], [4,5,6]]
print(m)
```

```python
print(m[0][1])
```

!!! warning Attention
Comme on travaille avec des listes de listes, on n'est pas sûr d'avoir des matrices valides. On peut avoir plusieurs lignes qui n'ont pas la même taille.\
Par exemple : 
`m = [[1,2],[3,4,5]]` fonctionne en python mais n'est pas une matrice.
!!!
!!! question Question 1 
Ecrire une fonction `est_matrice(m)` qui renvoie `True` ssi `m` est une matrice.
!!!

```python
def est_matrice(m):
    assert m != None and len(m) > 0, "m doit être non vide"
    
    nb_col = len(m[0])
    for i in range(len(m)):
        if nb_col != len(m[i]):
            return False
    return True
```

```python
print(est_matrice([[1,2,3], [4,5,6]])) # True
```

```python
print(est_matrice([[1,2],[3,4,5]])) # False
```

!!! question Question 2
Ecrire une fonction `afficher_matrice(m)` qui affiche une matrice ligne par ligne.
!!!

```python
def afficher_matrice(m):
    assert est_matrice(m), "m doit être une matrice"
    for ligne in m:
        print(ligne)
```

```python
afficher_matrice(m)
# [1, 2, 3]
# [4, 5, 6]
```

!!! question Question 3
Ecrire une fonction `addition(m1, m2)` qui additionne deux matrice de taille $n*m$.\
Renvoyer `None` si les matrices ne peuvent pas être additionner.
!!!

```python
""" version avec initialisation liste vide dans laquelle on ajoute ligne par ligne """
def addition(m1, m2):
    # cas ou m1 ou m2 n'est pas une matrice ou si elles ne peuvent être additionner
    if not est_matrice(m1) or not est_matrice(m2) or len(m1) != len(m2) or len(m1[0]) != len(m2[0]):
        return None
    
    res = []
    for i in range(len(m1)):
        ligne = []
        for j in range(len(m1[0])):
            ligne.append(m1[i][j] + m2[i][j])
        res.append(ligne)
    return res
```

```python
""" version avec initialisation matrice de la bonne taille avec que des 0 """
def addition(m1, m2):
    # cas ou m1 ou m2 n'est pas une matrice ou si elles ne peuvent être additionner
    if not est_matrice(m1) or not est_matrice(m2) or len(m1) != len(m2) or len(m1[0]) != len(m2[0]):
        return None
    
    res = [[0 for _ in range(len(m1[0]))] for _ in range(len(m1))]
    for i in range(len(m1)):
        for j in range(len(m1[0])):
            res[i][j] = m1[i][j] + m2[i][j]
    return res
```

```python
# test addition
m1 = [[1,2,3],[4,5,6]]
m2 = [[1,4,2],[4,6,1]]
afficher_matrice(addition(m1, m2))
```

```python
# test taille différentes
m1 = [[1,2,3],[4,5,6]]
m2 = [[1,4],[4,6]]
print(addition(m1, m2)) # None
```

```python
# test pas vrai matrice
m1 = [[1,2,3],[4,5]]
m2 = [[1,4,3],[4,6]]
print(addition(m1, m2)) # None
```

Érire une fonction `transpose(m)` qui renvoie la transposé la matrice `m`.

```python
""" version avec initialisation liste vide dans laquelle on ajoute ligne par ligne """
def transpose(m):
    assert est_matrice(m), "m doit être une matrice"
    res = []
    for j in range(len(m[0])): # colonne
        ligne = []
        for i in range(len(m)): # puis ligne
            ligne.append(m[i][j])
        res.append(ligne)
    return res
        
```

```python
""" version avec initialisation matrice de la bonne taille avec que des 0 """
def transpose(m):
    assert est_matrice(m), "m doit être une matrice"
    
    res = [[0 for _ in range(len(m))] for _ in range(len(m[0]))]
    for i in range(len(m)): 
        for j in range(len(m[0])): 
            res[j][i] = m[i][j]
    return res
        
```

```python
""" version compact """
def transpose(m):
    assert est_matrice(m), "m doit être une matrice"
    
    return [[m[j][i] for j in range(len(m))] for i in range(len(m[0]))]
```

```python
m = [[1,2,3,4],[5,6,7,8]]
afficher_matrice(transpose(m))
```

!!! question Question 5
Ecrire une fonction `multiplication(m1, m2)` qui multiplie deux matrices, `m1` de taille $n*p$ et `m2` de taille $p*q$.\
Renvoyer `None` si les matrices ne peuvent pas être multipliées.
!!!
!!! info Indice
$$ m1 =
\begin{pmatrix}
00 & 01 & 02 \\
10 & 11 & 12
\end{pmatrix}
\\
 m2 =
\begin{pmatrix}
00 & 01 \\
10 & 11 \\
20 & 21
\end{pmatrix}
\\
m1*m2 = 
\begin{pmatrix}
00*00 + 01*10 + 02*20 & 00*10 + 01*11 + 02*21 \\
10*00 + 11*10 + 12*20 & ...
\end{pmatrix}
$$
!!!

```python
""" version avec initialisation matrice de la bonne taille avec que des 0 """
def multiplication(m1, m2):
    # cas ou m1 ou m2 n'est pas une matrice ou si elles ne peuvent être multiplier
    if not est_matrice(m1) or not est_matrice(m2) or len(m1[0]) != len(m2):
        return None
    
    n = len(m1)
    p = len(m1[0])
    q = len(m2[0])
    
    res = [[0 for _ in range(q)] for _ in range(n)]
    
    for i in range(n):
        for j in range(q):
            for k in range(p):
                res[i][j] += m1[i][k] * m2[k][j]
    return res
```

```python
""" version avec initialisation liste vide dans laquelle on ajoute ligne par ligne """
def multiplication(m1, m2):
    # cas ou m1 ou m2 n'est pas une matrice ou si elles ne peuvent être multiplier
    if not est_matrice(m1) or not est_matrice(m2) or len(m1[0]) != len(m2):
        return None
    
    n = len(m1)
    p = len(m1[0])
    q = len(m2[0])
    
    res = []
    for i in range(n):
        ligne = []
        for j in range(q):
            element = 0
            for k in range(p):
                element += m1[i][k] * m2[k][j]
            ligne.append(element)
        res.append(ligne)
    return res
```

## Exercice 3 : Deux plus proches valeurs dans une liste de nombres

!!! question Question 1 
Ecrire une fonction `deux_plus_proches_valeurs(nombres)` qui, étant donnée une liste de nombres contenant au moins deux éléments, renvoie les deux valeurs les plus proches.\
`abs(x)` : renvoie la valeur absolue de x 
!!!

```python
def deux_plus_proches_valeurs(nombres):
    assert len(nombres)>1, "nombres doit contenir au moins 2 éléments"
    
    # initialisation avec les 2 premiers éléments
    indices = (nombres[0], nombres[1])
    ecart = abs(nombres[0] - nombres[1])
    
    for i in range(len(nombres)):
        for j in range(i+1, len(nombres)):
            ecart_i_j = abs(nombres[i] - nombres[j])
            if ecart_i_j < ecart :
                ecart = ecart_i_j
                res = (nombres[i], nombres[j])
    return res
                
    
```

```python
print(deux_plus_proches_valeurs([-1,3,5,2,6,8,11])) 
# renvoie (3,2) ou (2,3) ou (5,6) ou (6,5)
```

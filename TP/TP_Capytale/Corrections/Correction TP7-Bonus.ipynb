Avec l'aide de l'exemple ci-dessous venant de la documentation de Matplotlib 
https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_stacked.html#sphx-glr-gallery-lines-bars-and-markers-bar-stacked-py

Afficher un graphique en bar qui compare par année les consomations énergétiques des bâtiments tertiaires par activités. (Donc chaque bar correspond à une année et est découper en plusieurs couleurs qui représentes les secteurs d'activités.
!!!

````python
# EXEMPLE DE GRAPHIQUE BARPLOT AVEC COMMENTAIRE EN PLUS
import matplotlib.pyplot as plt
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = [
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
] # liste d'espèce (qui vont être une bar par espèce)

weight_counts = {
    "Below": [70, 31, 58],
    "Above": [82, 37, 66],
} #nos données sous forme d'un dictionnaire

width = 0.5 # épaisseur de nos bars

fig, ax = plt.subplots() # permet de créer plusieurs graphique (ici bar) sur une même figure

bottom = np.zeros(3)    # initialise un array numpy de 3 zéro, (on pourrait utiliser une liste) 
                        # qui va servir à définir la base de nos bars 

# creation des bars par catégories 
# en gros ici on a 2 tours de boucles, un qui mets le bleu des bars et un qui met le orange
for position, weight_count in weight_counts.items():
    # ajout des bars avec ax.bar()
    p = ax.bar(species, weight_count, width, label=position, bottom=bottom) 
    # on augmente les bases de nos bars (pour mettre les oranges au dessus des bleus.)
    bottom += weight_count


ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show() # afichage
````

!!! info 
Dans cet exercice, un mot binaire sera codé sous la forme d'une chaine de caractères, par exemple `'1010'` représente l'entier 10.

!!! question Question 1
Écrire une fonction `renverse(mot: str) -> str` qui inverse l'ordre des bits dans un mot binaire.
!!!

```python
def renverse(mot):
    return mot[::-1]
```

```python
print(renverse('1000101')) #1010001
```

!!! question Question 2
Écrire une fonction `bin_to_int(n_bin : str) -> int` qui transfome une représentation binaire vers l'entier naturel correspondant.
!!!

```python
def bin_to_int(n_bin):
    res = 0
    n_bin = renverse(n_bin)
    for i in range(len(n_bin)):
        if n_bin[i] == '1':
            res += 2**i
    return res
```

```python
# version sans reverse
def bin_to_int(n_bin):
    i = 0 
    res = 0
    while i < len(n_bin) :
        if n_bin[len(n_bin)-i-1] == '1':
            res += 2**i
        i += 1
    return res
```

```python
# autre version sans reverse
def bin_to_int(n_bin):
    i = len(n_bin) - 1 #dernier caractère
    res = 0
    while i >= 0 :
        if n_bin[i] == '1':
            res += 2**(len(n_bin)-i-1)
        i -= 1
    return res
```

```python
print(bin_to_int('1100100')) #100
```

!!! question Question 3
Écrire une fonction `int_to_bin(n_int: int) -> str` qui transforme un entier naturel en ça resprésentation binaire.
!!!

```python
def int_to_bin(n_int):
    res = ''
    while n_int != 0 :
        res += str(n_int%2)
        n_int = n_int//2
    return renverse(res)
```

```python
print(int_to_bin(100)) # 1100100
```

!!! question Question 1
Écrire une fonction `est_representable(n_int: int, n: int) -> bool` qui renvoie `True` si `n_int`, un entier relatif, peut être représenté en complément à 2 sur `n` bits, `False` sinon.
!!!

```python
def est_representable(n_int, n):
    val_max = 2**(n-1)-1
    val_min = -2**(n-1)
    return val_min <= n_int <= val_max
```

```python
print(est_representable(128, 8)) # False
```

```python
print(est_representable(127, 8)) # True
```

```python
print(est_representable(-128, 8)) # True
```

!!! question Question 2
Écrire une fonction `complement_a_1(n_bin: str) -> str` qui renvoie le complément à 1 (inversion des bits) du nombre binaire `n_bin`.
!!!

```python
def complement_a_1(n_bin):
    res = ''
    for bit in n_bin:
        if bit == '1':
            res += '0'
        else:
            res += '1'
    return res
```

```python
print(complement_a_1('1010110')) # 0101001
```

!!! question Question 3
Écrire une fonction `ajoute_1(n_bin: str) -> str`, qui ajoute 1 au nombre binaire `n_bin`. Attention, ici on veut garder le même nombre de bits (donc si dépassement il y a, ne pas le prendre en compte).

_Cette algorithme n'est pas si facile : je vous conseil de prendre une feuille pour dépiler tous les cas que vous pouvez trouver lors d'une addition de nombres binaires._
!!!

```python
def ajoute_1(n_bin):
    res = ''
    retenue = 1
    for i in range(len(n_bin)-1, -1, -1):
        if retenue == 0:
            res = n_bin[i] + res
        elif n_bin[i] == '0':
            retenue = 0
            res = '1' + res
        else:
            res = '0' + res
            
    return res
```

```python
def ajoute_1(n_bin):
    res = ''
    retenue = 1
    n_bin = renverse(n_bin)
    for bit in n_bin:
        if retenue == 0:
            res = bit + res
        elif bit == '0':
            retenue = 0
            res = '1' + res
        else:
            res = '0' + res
            
    return res
```

```python
print(ajoute_1('0000')) # 0001
```

```python
print(ajoute_1('11011011')) # 11011100
```

```python
print(ajoute_1('1111')) # 0000
```

!!! question Question 4
Écrire une fonction `int_to_ca2(n_int: int, n: int) -> str` qui convertie un entier relatif en ça représentation en complément à 2 sur `n` bits. 
Votre fonction renverra une erreur si cet entier n'est pas représentable sur `n` bits.
!!!

```python
def int_to_ca2(n_int, n):
    assert est_representable(n_int, n), "Le nombre de bit n'est pas suffisant"
    
    # on récupère la représentation la valeur absolue de n_int
    x = int_to_bin(abs(n_int))
    
    # on complète par des 0 pour aller jusqu'à n bits 
    for i in range(n-len(x)):
            x = '0' + x
            
    # si positif, on renvoie x
    if n_int >= 0 :
        return  x
    
    # sinon on fait le complément à 1
    x = complement_a_1(x)
    # puis on ajoute 1
    return ajoute_1(x)
```

```python
print(int_to_ca2(5, 8)) # 00000101
```

```python
print(int_to_ca2(-5, 8)) # 11111011
```

```python
print(int_to_ca2(-36, 8)) # 11011100
```

```python
print(int_to_ca2(-135, 8)) # Erreur 
```

!!! question Question 5
Écrire une fonction `ca2_to_int(n_ca2: str) -> int` qui convertie un binaire complément à 2 en entier.
!!!

```python
def ca2_to_int(n_ca2):
    if n_ca2[0] == '0' :
        return bin_to_int(n_ca2)
    n_ca2 = complement_a_1(n_ca2)
    n_ca2 = ajoute_1(n_ca2)
    return -bin_to_int(n_ca2)
```

```python
print(ca2_to_int('11001011')) # -53
```

```python
print(ca2_to_int('01001011')) # 75
```

```python
print(ca2_to_int(int_to_ca2(-67, 8))) # -67
```

!!! question Question 6
Écrire une fonction `addition(n_ca2_1: str, n_ca2_2: str) -> str` qui fait l'addition entre 2 nombres complément à 2. Vous vérifirez que les deux représentations sont sur le même nombre de bits.
!!!

```python
def addition(n_ca2_1, n_ca2_2):
    assert len(n_ca2_1) == len(n_ca2_2), "les 2 complément à 2 doivent avoir la même taille" 
    retenue = 0
    res = ''
    n_ca2_1 = renverse(n_ca2_1)
    n_ca2_2 = renverse(n_ca2_2)
    for i in range(len(n_ca2_1)):
        tot = retenue + int(n_ca2_1[i]) + int(n_ca2_2[i]) # on additionne les 2 bits et la retenue
        # puis on regarde tout les cas possible 
        if tot == 0:
            res = '0' + res
        elif tot == 1:
            retenue = 0
            res = '1' + res
        elif tot == 2:
            retenue = 1
            res = '0' + res
        else:
            res = '1' + res
    return res
            
```

```python
print(addition('00000101','11111011')) #00000000
```

## Exercice 4 Bonus : Récursif

!!! question Question 1
Écrire une fonction récursive `bin_to_int_1(n_bin)` qui renvoie l'entier positif associé à la représentation binaire `n_bin`. En faisant la somme des puissances de 2 de **gauche à droite**.
!!!

```python
def bin_to_int_1(n_bin):
    if len(n_bin) == 0:
        return 0
    puis = len(n_bin) - 1
    bit = int(n_bin[0])
    return bit * 2**puis + bin_to_int_1(n_bin[1:]) 
```

```python
print(bin_to_int_1('0101010101'))
```

!!! question Question 2
Écrire une fonction récursive `bin_to_int_2(n_bin)` qui renvoie l'entier positif associé à la représentation binaire `n_bin`. En faisant la somme des puissances de 2 de **droite à gauche**.
!!!

```python
def bin_to_int_2(n_bin, puis = 0):
    if len(n_bin) == 0:
        return 0
    bit = int(n_bin[-1])
    return bit * 2**puis + bin_to_int_2(n_bin[:-1], puis+1) 
```

```python
print(bin_to_int_2('0101010101'))
```

!!! question Question 3
Écrire une fonction récursive `bin_to_int_3(n_bin)` qui renvoie l'entier positif associé à la représentation binaire `n_bin`. \
On exploite la méthode suivante :
- $1001_{(2)} = 100_{(2)} * 2_{(10)} + 1 = 100_{(2)} * 2_{(10)} + 1_{(10)} $
- $100_{(2)} = 10_{(2)} * 2_{(10)} + 0_{(2)} = 10_{(2)} * 2_{(10)} + 0_{(10)}$
- $10_{(2)} = 1_{(2)} * 2_{(10)} + 0_{(2)} = 1_{(2)} * 2_{(10)} + 0_{(10)}$
- $1_{(2)} = 1_{(10)}$
!!!

```python
def bin_to_int_3(n_bin):
    if len(n_bin) == 1:
        return int(n_bin)
    bit = int(n_bin[-1])
    return bit + 2*bin_to_int_3(n_bin[:-1])
```

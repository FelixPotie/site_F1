Soit la suite définie par : $\begin{cases}
    u_0 = 1 \\     
    \forall n \in  \mathbb{N}, u_{n+1} = \frac{1}{2} u_n +1
  \end{cases}
  $\
Écrire une fonction `suite_un(n)` récursive qui reçoit un entier $n$ et renvoie la valeur de $u_n$ de la suite.
!!!


```python
def suite_un(n):
    if n == 0:
        return 1
    else:
        return 0.5 * suite_un(n-1) + 1
```

## Exercice 2 : Exponentiation rapide
!!! info
On note $⌊n/2⌋$ la partie entière de $n/2$.\
Soient les 4 fonctions suivantes permettant toutes de calculer $a^n, n \in \mathbb{N}$ :

- `exp_1(a,n)` : $a^n = a^n$

- `exp_2(a,n)` : $a^n = a^{n-1} * a$

- `exp_3(a,n)` : $a^n = a^{⌊n/2⌋} * a^{n-⌊n/2⌋}$

- `exp_rapide(a,n)` réalise un calcul en exploitant : $a^n = \begin{cases}
    (a^{⌊n/2⌋})²       & \quad \text{si } n \text{ est pair}\\
    (a^{⌊n/2⌋})² * a  & \quad \text{sinon} 
  \end{cases}
  $
!!!

!!! question Question 1
Écrire la fonction `exp_1(a,n)` qui fait $a^n$ avec la façon de faire de base de python.
!!!

```python
def exp_1(a,n):
    return a ** n

print(exp_1(2,12))
```

!!! question Question 2
Écrire la fonction `exp_2(a,n)` qui fait $a^n = a^{n-1} * a$ de façon **itérative** (boucle for ou while).
!!!

```python
def exp_2(a,n):
    res = 1
    for _ in range(n):
        res*=a
    return res
print(exp_2(2,12))
```

!!! question Question 4
Écrire la fonction **récursive**`exp_3(a,n)` qui fait $a^n = a^{⌊n/2⌋} * a^{n-⌊n/2⌋}$.
!!!

```python
def exp_3(a,n):
    if n == 0:
        return 1
    if n == 1 :
        return a
    k = n//2
    return exp_3(a,k) * exp_3(a,n-k)
print(exp_3(2,12))
```

!!! question Question 5
Écrire la fonction **récursive** `exp_rapide_r(a,n)` qui réalise un calcul en exploitant : $a^n = \begin{cases}
    (a^{⌊n/2⌋})²       & \quad \text{si } n \text{ est pair}\\
    (a^{⌊n/2⌋})² * a  & \quad \text{sinon} 
  \end{cases}
  $
!!!

```python
def exp_rapide_r(a,n):
    if n == 0 :
        return 1
    tmp = exp_rapide_r(a, n//2)
    if n % 2 == 0 :
        return tmp * tmp
    else:
        return tmp * tmp * a
print(exp_rapide_r(2,12))     
```

!!! question Question 6
Écrire la fonction **itérative** `exp_rapide(a,n)` qui réalise un calcul en exploitant : $a^n = \begin{cases}
    (a^{⌊n/2⌋})²       & \quad \text{si } n \text{ est pair}\\
    (a^{⌊n/2⌋})² * a  & \quad \text{sinon} 
  \end{cases}
  $
\
L'algorithme est le suivant : 
- on initialise une variable `res` à 1
- tant que `n` n'est pas égale à 0 :
    - si `n` est impaire `res` devient `res * a`
    - on remplace `a` par `a * a` 
    - on remplace `n` par $⌊n/2⌋$
!!!

```python
def exp_rapide(a,n):
    res = 1
    while n != 0:
        if n % 2 == 1 :
            res *= a
        a *= a
        n = n//2
    return res
print(exp_rapide(2,12))
```

!!! question Question 1
Sans regarder la solution dans le cours, écrivez un algorithme de recherche dichotomique **récursif**. O cherche donc un élément dans une liste trié, on renverra l'indice de l'élément s'il est présent, None sinon.

La compléxité de votre fonction devra être en $O(log(n))$ où n est la taille de laliste passé en paramètre. Attention il n'est donc pas possible de faire celà avec les slices comme vu en cours.
!!!

```python
def rech_dicho(liste, x, debut, fin):
    if debut > fin:
        return None  # x non trouvée

    milieu = (debut + fin) // 2

    if liste[milieu] == x:
        return milieu
    elif liste[milieu] < x:
        return rech_dicho(liste, x, milieu+1, fin)
    else:
        return rech_dicho(liste, x, debut, milieu-1)

def recherche_dichotomique(liste, x):
    return rech_dicho(liste, x, 0, len(liste) - 1)
```

```python
print(recherche_dichotomique([i for i in range(100)], 25))
```

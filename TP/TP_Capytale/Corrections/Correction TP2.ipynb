## Exercice 1 : Nombres parfaits

L'objectif de cet exercice est de trouver les nombres parfaits.
!!! info Nombre parfait
En arithmétique, un nombre parfait est un entier naturel égal à la somme de ses diviseurs positifs sauf lui-même (diviseurs stricts). \
Ainsi 6 est un nombre parfait car ses diviseurs positifs sont 1, 2, 3 et 6, et il vérifie bien 6 = 1 + 2 + 3.\
Mais 12 n'est pas un nombre parfait car ses diviseurs positifs sont 1, 2, 3, 4, 6 et 12, mais 1 + 2 + 3 + 4 + 6 = 16.
!!!
Pour arriver à ce résultat on va effectuer des fonctions intermédiaires.

!!! question Question 1
Ecrivez une fonction `liste_diviseurs(n)` qui, étant donné un entier naturel `n`, renvoie la liste des diviseurs positifs de `n` qui sont strictement inférieurs à `n`.

Vous pourrez vous aider des instructions : 
- `a // b` : qui renvoie le quotien de la division de a par b
- `a % b` : qui renvoie le reste de la division de a par b
!!!

```python
def liste_diviseurs(n):
    diviseurs = []
    for i in range(1,n//2+1): # on sait qu'on n'a pas de diviseur plus grand que n/2
        if n%i == 0:
            diviseurs.append(i)
    return diviseurs
```

```python
print(liste_diviseurs(50)) 
# doit renvoyer [1,2,5,10,25]
```

!!! question Question 2
Ecrivez une fonction `somme_1(t)` qui, étant donnée une liste de nombres `t`, renvoie la somme de ses éléments en itérant sur les **indices** de `t`.
!!!

```python
def somme_1(t):
    s = 0
    for i in range(len(t)):
        s += t[i]
    return s
```

```python
print(somme_1([]))
print(somme_1([1,3,5,3]))
print(somme_1([1,-3,5,-4]))
```

!!! question Question 3
Ecrivez une fonction `somme_2(t)` équivalente mais en itérant sur les **éléments** de `t`.
!!!


```python
def somme_2(t):
    s = 0
    for element in t:
        s += element
    return s
```

```python
print(somme_2([]))
print(somme_2([1,3,5,3]))
print(somme_2([1,-3,5,-4]))
```

## Exercice 2 : Recherche séquentielle
Recherche séquentielle d’un élément dans une liste ou dans une chaîne de caractère.\
Pour effectuer une recherche séquentielle d'un élément x dans une liste ou une chaîne de caractère t on procède de la manière suivante :
- On parcourt les éléments de t un par un
- Si l'élément en cours correspond à x on renvoie `Vrai`
- Une fois tous les éléments testés, si aucun ne correspondait à x, on renvoie `Faux`

!!! question Question 1 
Ecrire une fonction `appartient(x,t)` qui renvoie `True` si x (nombre ou chaine de caractères d’une seule lettre) se trouve dans t (liste ou chaine de caractères) et `False` sinon à l'aide d'une recherche séquentielle
!!!

```python
def appartient(x, t):
    for element in t:
        if x == element:
            return True
    return False
```

```python
print(appartient(42,[12, 17, 42, 5]))
# doit renvoyer True
```

```python
print(appartient(24,[12, 17, 42, 5]))
# doit renvoyer False
```

```python
print(appartient('s','testons'))
# doit renvoyer True
```

Ecrire une fonction `liste_positions(x,t)` qui renvoie la liste (éventuellement vide) de toutes les positions de l'élément `x` dans `t`.
!!!

```python
def liste_positions(x, t):
    positions = []
    for i in range(len(t)):
        if x == t[i]:
            positions.append(i)
    return positions
```

## Exercice 3 : Détermination des maximums d'une liste de nombres

On cherche dans un premier temps à déterminer le minimum d'une liste.

Le principe est le suivant: on utilise une variable afin de stocker le maximum provisoire. Voici les étapes:
- Si la liste est vide, on renvoie `None`
- On initialise une variable `max` à la valeur du 1er élément de la liste
- On parcourt les éléments de la liste à l’aide d’une boucle
    - Si l’élément en cours est plus grand que `max` on met `max` à jour
    - Sinon, on ne fait rien
- Une fois arrivé au bout de la liste, on renvoie `max`

!!! question Question 1
Ecrire la fonction `maximum(liste)`
!!!

```python
def maximum(liste):
    # cas liste vide
    if len(liste) == 0:
        return None
    # cas liste non vide
    max_val = liste[0]
    for element in liste:
        if element > max_val :
            max_val = element
    return max_val
```

```python
print(maximum([])) # None
```

```python
print(maximum([3,4,2,8])) # 8
```

!!! question Question 4 
Ecrire une fonction `second_max(liste)` qui renvoie le second maximum de `liste` en ne parcourant `liste` qu’une seule fois.
!!!

```python
def second_max(liste):
    assert len(liste) >= 2
    
    # on prend le 2eme min des 2 premiers éléments
    if liste[0] > liste[1]:
        max_1 = liste[0]
        max_2 = liste[1]
    else:
        max_1 = liste[1]
        max_2 = liste[0]
    
    # on parcourt la liste à partir du 3eme élément
    for i in range(2, len(liste)):
        # à chaque tour de boucle max_1 et max_2 sont les max et second max dans liste[0:i]
        if liste[i] > max_2:
            if liste[i] > max_1:
                max_2 = max_1
                max_1 = liste[i]
            else:
                max_2 = liste[i]
    return max_2
```

```python
print(second_max([]))
```

```python
print(second_max([3,2,7,1,5])) # 5
```

Ecrire une fonction `indice_second_max(liste)` qui renvoie une des positions du second maximum en ne parcourant `liste` qu’une seule fois.

```python
def indice_second_max(liste):
    assert len(liste) >= 2
    
    # on prend le 2eme min des 2 premiers éléments
    if liste[0] > liste[1]:
        max_1_i = 0
        max_2_i = 1
    else:
        max_1_i = 1
        max_2_i = 0
    
    # on parcourt la liste à partir du 3eme élément
    for i in range(2, len(liste)):
        # à chaque tour de boucle max_1_i et max_2_i sont les indices des max et 2eme max dans liste[0:i] 
        if liste[i] > liste[max_2_i]:
            if liste[i] > liste[max_1_i]:
                max_2_i = max_1_i
                max_1_i = i
            else:
                max_2_i = i
    return max_2_i
```

!!! question Question 6
Créer une fonction `n_max(liste,n)` renvoyant dans l’ordre, les `n` premiers maximum de `liste`. 
!!!

```python
# version avec l.remove(x) qui suppriler l'element x de la liste
def n_max(liste, n):
    assert n <= len(liste)
    res = []
    for i in range(n):
        max_i = max(liste)
        res.append(max_i)
        liste.remove(max_i)
    return res
```

```python
print(n_max([3,5,8,6,5,3,5,9,10,3,12],4)) # [12,10,9,8]
```

```python
t = [3,5,8,6,5,3,5,9,10,3,12]
print(n_max(t, len(t)))
```

```python
# version bonus sans liste.remove(x)

def maximum_inferieur_a_x(liste, x):
    """renvoie la valeur du maximum de liste qui est inférieur à x"""
    max_val = None
    for element in liste:
        if element < x :
            if not max_val or element > max_val: # revient à faire if max_val == None or or element > max_val
                max_val = element 
    return max_val

def nb_occ(liste, x):
    """renvoie le nombre de x dans liste"""
    cpt = 0
    for e in liste:
        if e == x: 
            cpt += 1
    return cpt
                
def n_max_2(liste, n):
    """renvoie une liste contenant les n plus grands max de liste (dans l'ordre décroissant)"""
    assert n <= len(liste) and n > 0
    res = []
    # on ajoute le 1er maximum pour pouvoir utiliser notre fonction maximum_inferieur_a_x()
    res.append(max(liste))
    for i in range(1,n):
        nouveau_max = maximum_inferieur_a_x(liste, res[len(res)-1])
        for j in range(nb_occ(liste,nouveau_max)):
            res.append(nouveau_max)
    return res
```

```python
print(n_max_2([3,5,8,6,5,3,5,9,10,3,12],4)) # [12,10,9,8]
```

```python
t = [3,5,8,6,5,3,5,9,10,3,12]
print(n_max_2(t, len(t)))
```

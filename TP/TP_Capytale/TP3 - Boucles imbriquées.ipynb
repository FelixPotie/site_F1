## Exercice 2 : Addition et multiplication de matrice

!!! info Matrice 
On rappelle qu'une matrice est une famille de nombres $(a_{i,j})$ qu'on écrit souvent sous la forme d'un tableau : 
\begin{pmatrix}
a_{11} & a_{12} & \cdots & a_{1p} \\ 
a_{21} & a_{22} & \cdots & a_{2p} \\ 
\vdots & \vdots &  & \vdots \\ 
a_{n1} & a_{n2} & \cdots & a_{np}
\end{pmatrix}

En python on utilisera des **listes de listes** pour représenter des matrices, mais attention contrairement aux habitudes mathématiques, **l'indexation en python commencera à 0**.
!!!
!!! example Exemple de matrice
On peut initialiser la matrice : 
$$ m =
\begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{pmatrix}
$$
en faisant : 
`m = [[1,2,3], [4,5,6]]`

On peut accéder à la valeur située à la ligne `i` et la colonne `j` en faisant `m[i][j]`. 

`m[i]` renverra la ligne `i`.

Attention les indices commencent toujours à `0`.
!!!
!!! info Taille de matrice
`len(m)` renvoie le nombre de ligne\
`len(m[i])` renvoie le nombre d'éléments de la ligne `i`, qui représente donc le nombre de colonnes de la matrice
!!!

```python
m = [[1,2,3], [4,5,6]]
print(m)
```

```python
print(m[0][1])
print(m[1])
```

!!! warning Attention
Comme on travaille avec des listes de listes, on n'est pas sûr d'avoir des matrices valides. En effet, dans une liste de liste on peut avoir plusieurs lignes qui n'ont pas la même taille.\
Par exemple : 
`m = [[1,2],[3,4,5]]` fonctionne en python mais n'est pas une matrice.
!!!
